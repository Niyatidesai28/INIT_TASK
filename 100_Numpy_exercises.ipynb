#### 56. Generate a generic 2D Gaussian-like array (★★☆)

````python
n = 64
sigma = 10.0

x = np.linspace(-n/2, n/2, n)
y = np.linspace(-n/2, n/2, n)
X, Y = np.meshgrid(x, y)

G = np.exp(-(X**2 + Y**2) / (2 * sigma**2))

print(G.shape)
````

#### 57. How to randomly place p elements in a 2D array? (★★☆)
